# Úkol 1. Knihovny a moduly pro matematické výpočty

```python
#Klasický python
def skalarni_soucin_py(pole1,pole2, i=0, vysledek = 0):
    start = time.time()
    while i < len(pole1):
        vysledek = pole1[i]*pole2[i] + vysledek
        i = i+1
    konec = time.time()
    print(f"{vysledek}, Čas: {round(konec-start,10)}")

#Numpy
def skalarni_soucin(pole1, pole2):
    start = time.time()
    vysledek = numpy.dot(pole1, pole2)
    konec = time.time()
    print(f"{vysledek}, Čas: {round(konec-start,10)}")

pole1 = numpy.arange(0, 100000)
pole2 = numpy.arange(0, 200000,2)

skalarni_soucin_py(pole1, pole2)
skalarni_soucin(pole1, pole2)
warnings.filterwarnings("ignore")
```

```python
#Klasický python
def nasob_py(od,do, j=0, listik = [], vysledek = 1):
    start = time.time()
    for i in range(od,do):
        listik.append(i)
    while j < len(listik):
        vysledek = listik[j] * vysledek
        j+=1
    konec = time.time()
    print(f"{vysledek}, Čas: {round(konec-start,10)}")

#Math
def nasob(od, do, listik=[]):
    start = time.time()
    for i in range(od,do):
        listik.append(i)
    vysledek = math.prod(listik)
    konec = time.time()
    print(f"{vysledek}, Čas: {round(konec-start,10)}")

nasob(1,1500)
nasob_py(1,1500)
```

```python
def f(x): 
    return -5*x**2 + 6*x + 7
#Klasický Python
def integral_py(a,b,dx):
    start = time.time()
    n = int((b-a)//dx)+1
    integral = f(a) + f(b)
    for i in range(1, n):
        integral += 2*f(a+i*dx)
    vysledek = integral * dx/2
    konec = time.time()
    print(f"{vysledek}, Čas: {round(konec-start,10)}") 

#Numpy
def integral_numpy(a,b,dx):
    start = time.time()
    x = numpy.arange(a, b+dx, dx)
    y = f(x)
    konec = time.time()
    vysledek = numpy.trapz(y, dx=dx)
    print(f"{vysledek}, Čas: {round(konec-start,10)}") 

integral_py(100000,565655,0.1)
integral_numpy(100000,565655,0.1)
```

```python
#Klasický Python
def det_py(a, j=0, k=0, x=0, vysledek_1 = 0, vysledek_2 = 0):
    start = time.time()
    for i in range(0,len(a[0])):
        x=0
        prvni = a[x][j] * a[x+1][j+1] * a[x+2][j+2]
        j = j - 2
        if j == -4:
            j = j+3
        vysledek_1 = vysledek_1 + prvni
    for i in range(0, len(a[0])):
        x=0
        prvni = a[x][k-1] * a[x+1][k-2] * a[x+2][k]
        k = k - 1
        if k == -2:
            k = k+3
        vysledek_2 = vysledek_2 + prvni 
    vysledek = vysledek_1-vysledek_2
    konec = time.time()
    print(f"{vysledek}, Čas: {round(konec-start,10)}") 

#Numpy
def det_numpy(a):
    start = time.time()
    vysledek = numpy.linalg.det(a)
    konec = time.time()
    print(f"{vysledek}, Čas: {round(konec-start,10)}") 
    
a = numpy.array([[41, 16, 4544], [223, 124, 14], [113, 23, 31]])
det_py(a)
det_numpy(a)
```

```python

def factorial_math(x):
    start = time.time()
    vysledek = math.factorial(x)
    konec = time.time()
    print(f"{vysledek}, Čas: {round(konec-start,10)}")

#vlastni reseni
def factorial(x, vysledek = 1):
    start = time.time()
    for i in range(1, x+1):
        vysledek = vysledek * i
    konec = time.time()
    print(f"{vysledek}, Čas: {round(konec-start,10)}")

factorial_math(1000)
factorial(1000)
```

# Úkol 2. Vizualizace dat

```python
def pyplot_make(file):
    font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
    
    df = pandas.read_csv(file)
    x = ["2010", "2015", "2020", "2022"]
    y = df["2010 Population"][87], df["2015 Population"][87], df["2020 Population"][87], df["2022 Population"][87]
    fig, ax = pyplot.subplots(figsize=(6,4))
    pyplot.plot(x,y, marker="o")
    pyplot.axis([0,3,10450000,10550000])
    pyplot.xlabel("Populace", font)
    pyplot.ylabel("10 Milionů", font)
    pyplot.title("Křivka populace ČR", font)
    for index in range(len(x)):
        ax.text(x[index], y[index], y[index], size=12)

pyplot_make('population.csv')
```

```python
def pybar_make(file, i=0, k=0, x =[], y=[]):
    font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
    
    df = pandas.read_csv(file)
    df = df.sort_values(by=['World Population Percentage'], ascending=False).head(5)
    for i in df['Country']:
        x.append(i)
    for k in df['World Population Percentage']:
        y.append(k)
    fig, ax = pyplot.subplots(figsize=(6,4))
    pyplot.bar(x,y)
    pyplot.axis([-0.5,5,0,20])
    pyplot.xlabel("Stát", font)
    pyplot.ylabel("Procenta", font)
    pyplot.title("5 Nejlidnatějších států", font)
    for index in range(len(x)):
        ax.text(x[index], y[index], y[index], size=12)

pybar_make('population.csv')
```

# Úkol 9. Integrace funkce jedné proměnné

```python
#funkce
def polynomial_f(x):
    return x**3 - 3*x + 7

def harmonic_f(x):
    return 2*numpy.sin(3*x) 

def logarithm_f(x):
    return numpy.log(6*x) + (2/3)

#funkce vypoctu 
def riemann(fce, a, b):
    return scipy.integrate.quadrature(fce, a ,b)

def simpson(funkce, c, d, h=0.01):
    return scipy.integrate.simpson(funkce(numpy.arange(c, d+h, h)), numpy.arange(c, d+h, h))

def romberg(fce, e, f):
    return scipy.integrate.romberg(fce, e, f)
 
#vysledky
print("Polynomicka funkce")
print(f"Riemann {riemann(polynomial_f, 1, 2)[0]}")
print(f"Simpson {simpson(polynomial_f, 1, 2)}")
print(f"Romberg {romberg(polynomial_f, 1, 2)}")

print("\nHarmonicka funkce")
print(f"Riemann{riemann(harmonic_f, 1, 2)[0]}")
print(f"Simpson {simpson(harmonic_f, 1, 2)}")
print(f"Romberg {romberg(harmonic_f, 1, 2)}")

print("\nLogaritmicka funkce")
print(f"Riemann {riemann(logarithm_f, 1, 2)[0]}")
print(f"Simpson {simpson(logarithm_f, 1, 2)}")
print(f"Romberg  {romberg(logarithm_f, 1, 2)}")
```

# Úkol 6. Generování náhodných čísel a testování generátorů

```python
def generate_seed_from_mouse_movement(samples=2):
    movements = []
    for _ in range(samples):
        x, y = pyautogui.position()
        movements.append((x, y))
        pyautogui.move(random.randint(-10, 10), random.randint(-10, 10), duration=0.1)
    seed = hash(tuple(movements))
    return seed

for i in range(0,10):
    s = generate_seed_from_mouse_movement()
    print("Vygenerované semínko:", s)








```

# Úkol 5. Hledání kořenů rovnice

```python
p1 = lambda x : x**3 - 1
p2 = lambda x : -numpy.sin(3*x) + x
p3 = lambda x : numpy.log(x**2) - 2

#Uzavrena metoda - bisekce, puleni intervalu
def uzavrena(fce, min0, max0, presnost=0.0001, opakovani =0):
    min, max = min0, max0
    while abs(max-min) > presnost:
        koren = (max + min)/2
        if fce(min)*fce(max) < 0:
            max = koren
        else:
            min = koren
        opakovani += 1
    print(f"Opakovani pro presnost 0.001: {opakovani}")
    return koren

def vysledekUM(fce, min, max, koren, nazev):
    print(f"Koren: {koren}")
    print(f"Funkcni hodnota: {fce(koren)}")
    pyplot.plot(numpy.linspace(min, max), fce(numpy.linspace(min, max)), "r")
    pyplot.plot(koren, fce(koren), "ro")
    pyplot.title(nazev)
    pyplot.show()

vysledekUM(p1, 0.1, 10, uzavrena(p1, 0.1, 5), "x**3 - 1")
vysledekUM(p2, 0.1, 10, uzavrena(p2, 0.1, 1), "-sin(3*x) + x")
vysledekUM(p3, 0.1, 10, uzavrena(p3, 1, 3), "log(x**2) - 2")


#otevrena metoda - Newtonova metoda tecen (pomoci derivace)
def otevrena(fce, bod, presnost=0.00001, opakovani = 0):
    while abs(fce(bod)) > presnost:
        koren = bod - (fce(bod) / scipy.misc.derivative(fce, bod))
        bod = koren
        opakovani += 1
    koren = bod
    print(f"Opakovani pro presnost 0.001: {opakovani}")
    return koren


def vysledekOM(fce, min, max, koren, nazev):
    print(f"Koren: {koren}")
    print(f"Funkcni hodnota: {fce(koren)}")
    pyplot.plot(numpy.linspace(min, max), fce(numpy.linspace(min, max)), "r")
    pyplot.plot(koren, fce(koren), "ro")
    pyplot.title(nazev)
    pyplot.show()

vysledekOM(p1, 0.1, 10, otevrena(p1, 2), "x**3 - 1")
vysledekOM(p2, 0.1, 10, otevrena(p2, 0.1), "-sin(3*x) + x")
vysledekOM(p3, 0.1, 10, otevrena(p3, 2), "log(x**2) - 2")
```
